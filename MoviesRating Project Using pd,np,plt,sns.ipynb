```python
# How to working with joint plots
import matplotlib as plt  # Visualization 
import seaborn as sns  # advanced Visualization
%matplotlib inline    # All the plot should  inside the line
import warnings
warnings.filterwarnings('ignore')
```

```python
j = sns.jointplot(data=movies, x='CriticRating', y='AudienceRating')
j
```

```python
j = sns.jointplot(data=movies, x='CriticRating', y='AudienceRating', kind='reg')
j
```

```python
j = sns.jointplot(data=movies, x='CriticRating', y='AudienceRating', kind='scatter')
j
```
